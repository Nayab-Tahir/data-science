## 3. Scatter Plot
A scatter plot is used to plot data points on a figure based on the horizontal and vertical axes. Scatter plots can be used to show the relationship between two variables. They can also show how data clusters in a dataset. Additionally, you can also use a third variable to determine the size or color of the points. 


In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid. Additonally, you can also use a third variable to determine the size or color of the points.

The [Iris flower dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set) provides sample measurements of sepals and petals for three species of flowers. The Iris dataset is included with the Seaborn library and can be loaded as a Pandas data frame.

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
# To handle URLError: <urlopen error [SSL: CERTIFICATE_VERIFY_FAILED] ... unable to get local issuer certificate>
import ssl
ssl._create_default_https_context = ssl._create_unverified_context



df_iris = sns.load_dataset("iris")
df_iris.head()
```

```python
sns.scatterplot(x=df_iris.sepal_length, y=df_iris.sepal_width);
```

Seaborn has in-built support for Pandas data frames. Instead of passing each column as a series, you can provide **column names** and use the `data` argument to specify a data frame.

```python
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data=df_iris);
```

### Adding Hues
- Notice how the points in the above plot seem to form distinct clusters with some outliers. 
- We can color the dots using the flower species as a `hue`, which can be another column having same length as the sepal_width and sepal_length column. For `hue`, we use the species argument as for each record is of specific species.
- We can also make the points larger using the `s` argument.

```python
sns.scatterplot(x = df_iris.sepal_length, y = df_iris.sepal_width, hue = df_iris.species, s=100);
```

Adding `hues` makes the plot more informative. We can immediately tell that Setosa flowers have a smaller sepal length but higher sepal widths. In contrast, the opposite is true for Virginica flowers. 

### Customizing Seaborn Figures
- Since Seaborn uses Matplotlib's plotting functions internally, we can use functions like `plt.figure` and `plt.title` to modify the figure, before the `sns.scatterplot()` method. So in this way you can mix and match thing among Matplotlib and Seaborn
- Moroever, `plt.subplots()` method can also be used to set the figure size, it is more verstile as it can be used for sub-plots (More on this later)

```python
#plt.figure(figsize=(12, 8))
fig, ax = plt.subplots(figsize=(12, 8))
plt.title('Sepal Dimensions')

sns.scatterplot(x = df_iris.sepal_length, 
                y = df_iris.sepal_width, 
                hue = df_iris.species,
                s=100);
```

## 4. Histogram
>- A histogram represents the distribution of a variable by creating bins (intervals) along the range of values and showing vertical bars to indicate the number of observations in each bin. 



A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.


>- The histogram is an important graph in statistics and data analysis. It can be used to help people quickly understand the distribution of data. 
>- In order to draw a histogram, we follow the steps outlined below:
    - Step 1: Bin the range of your data.
    - Step 2: Divide the entire range of values into their corresponding bins.
    - Step 3: Count how many values fall into each different bin

>- We can use the `sns.histplot()` method of Seaborn or the `plt.hist()` method of Matplotlib to create a histogram.

### a. Histogram of a Randomly Generated Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Use the np.random.normal() method to get a Normal Data Distribution


# Draw random samples from a normal (Gaussian) distribution.
arr = np.random.normal(size=1000)

# You can use the `sns.histplot() method of seaborn to visualize the distribution
sns.histplot(data=arr, kde=True);

# You can also use the `plt.hist()` method of Matplotlib to visualize the distribution
#plt.hist(arr);
```

### b. Histogram of `sepal_width` in the  `iris` Dataset
- Histogram is normally about understanding the distribution of a single variable
- Let us load the `iris` dataset and look at the distribution of sepal_width only

```python
# Load data into a Pandas dataframe
df_iris = sns.load_dataset("iris")
df_iris.sepal_width
```

```python
# Check statistical information about sepal_width 
df_iris.describe()
```

- Above table tells you that the minimum value of `sepal_width` is 2.0 and the maximum value is 4.4. The entire data is spread within this range.
- However, this doesn't tell us where the different values lie, i.e., it doesn't give answer to questions like
    - how many flowers have the exact `sepal_width` of 3.0
    - how many flowers have the `sepal_width` between 2.9 and 3.1
- By drawing a histogram of `sepal_width`, we can easily visualize the distribution and can give the answer to such questions.

```python
plt.title("Distribution of Sepal Width")
sns.histplot(data=df_iris.sepal_width);
#plt.hist(df.sepal_width);
```

We can immediately see that the sepal widths lie in the range 2.0 - 4.5, and around 35 values are in the range 2.9 - 3.1, which seems to be the most populous bin.

### c. Controlling the Numbers, Size of Bins/Intervals
- The bins intervals shown by default are arbitrary, and it is difficult to check out the distribution between a specific interval, e.g., between 2.0 and 2.2
- The solution to this is by controlling the number of bins or the size of each bin using the `bins` argument to the `sns.histplot()`.

```python
# Specifying the number of bins
sns.histplot(df_iris.sepal_width, bins=6);
#plt.hist(df.sepal_width, bins=6);
```

```python
# Specifying the Boundry points of each Bin
sns.histplot(df_iris.sepal_width, bins=np.arange(2, 5, 0.25));
#plt.hist(df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
# Bins of Unqual Sizes (1 to 3 and 3 to 4 and 4 to 4.5)
sns.histplot(df_iris.sepal_width, bins=[1, 3, 4, 4.5]);
#plt.hist(df.sepal_width, bins=[1, 3, 4, 4.5]);
```

### d. Multiple Histograms in one Figure
- Similar to line charts, we can draw multiple histograms in a single chart. We can reduce each histogram's opacity  so that one histogram's bars don't hide the others'.
- Let's draw separate histograms for each species of flowers.

```python
# Example: Draw two histograms, one for sepal_width of setosa and other for sepal_width of versicolor
setosa_df = df_iris[df_iris.species == 'setosa']
versicolor_df = df_iris[df_iris.species == 'versicolor']
setosa_df
#versicolor_df
```

```python
sns.histplot(setosa_df.sepal_width);
sns.histplot(versicolor_df.sepal_width);
```

**You see there is an overlap and the information is not very clear, set the opacity by using the alpha parameter to `sns.histplot()` method**

```python
sns.histplot(setosa_df.sepal_width, alpha=0.2);
sns.histplot(versicolor_df.sepal_width, alpha=0.4);
plt.legend(['Setosa', 'Versicolor']);
```

>This histogram now gives us more information like Setosa flowers have sepal_width on the larger side, while Versicolor has a smaller sepal width 

### e. Stacking Histograms 
- We can also stack multiple histograms on top of one another.**

```python
setosa_df = df_iris[df_iris.species == 'setosa']
versicolor_df = df_iris[df_iris.species == 'versicolor']
virginica_df = df_iris[df_iris.species == 'virginica']

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.title('Distribution of Sepal Width')

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

## 6. Heat Map

>- A heatmap is a useful chart that we can use to show the relationship between two variables. 
>- A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors. 
>- Let's look at a sample dataset included with Seaborn, called `flights`. 
>- The dataset contains following information about the monthly passenger footfall at an airport over 12 years

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
# To handle URLError: <urlopen error [SSL: CERTIFICATE_VERIFY_FAILED] ... unable to get local issuer certificate>
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

#Load the tips dataset
df_flights = sns.load_dataset("flights")
df_flights
```

**It contain 12 months data of years from 1949 to 1960. So a total of 12*12=144 records. Each record contains count of passengers in thousands, who visited the airport. So the first record says that in the month of January, 1949, a total of 112 thousand passengers visited the airport**

**We want to know the trend of passengers visiting the airport, is it increasing or decreasing**

**Let us draw a line plot to visualize this trend**

```python
plt.plot(df_flights.passengers);
```

**It shows an increasing trend, however, a nicer way to visualize is to use a heatmap chart**
**For that first we have to represent this data as a metrix having months as rows and year as columns**

```python
df_matrix = sns.load_dataset("flights").pivot("month", "year", "passengers")
df_matrix
```

- `df_matrix` is a matrix with one row for each month and one column for each year. Each intersection of a row and column tells us the count of passengers (in thousands) who visited the airport in that month of the year.
- Thus there are a total of 144 data values in a matrix shape
- We can use the `sns.heatmap()` method to visualize the footfall at the airport.

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(df_matrix);
```

Instead of numbers, we see the colours. The darker the colour represent a lower number. You can also observe the legend on the right side, which shows the passenger count and the corresponding colour in the heatmap. The brighter colors indicate a higher footfall at the airport.
By looking at the graph, we can infer two things: 
- The footfall at the airport in any given year tends to be the highest around July & August.
- The footfall at the airport in any given month tends to grow year by year.**

- We can also display the actual values in each block by specifying `annot=True` 
- We use `cmap` argument to change the color palette, and also the legend shows lighter colors indicate a lower count
- We have used `fmt` argument to display the in integers (default is float pointing exponent notation)


```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(df_matrix, annot=True, cmap='Blues', fmt='d');
```

## 8. Pie Chart
Pie charts are often used to highlight the proportion of categorical data in relation to the overall value of a dataset. Pie charts are shaped like a pie, and each category that takes up a part of the pie is shaped like a slice. The larger the slice, the greater the proportion of the overall values it represents.

Like with other plot types, the basic function in Matplotlib to plot a pie chart is `pie()`. Below are some of the important parameters we may need:

- `x`: The list of data we want to present for each Wedge.
- `explode`: Set as None or as a list of values that are the same length as x. Each item in this list indicates how far each wedge deviates from the center of the circle
- `labels`: A sequence of string that provides labels for each wedge.
- `colors`: Set as None or as a list of colors that are the same length as x.
- `shadow`: Set as True or False. the default values is False. Setting true adds a shadow effect.
- `counterclock`: Set as True or False. The default value is True. Used to specify the direction of the fractions.
- `autopct`: Set as none or as a string to label wedges with their numeric values. The label will be placed inside the wedge.

### a. Draw pie chart

### c. Adding explode to the pie
Sometimes, we want to focus on a specific category, such as the one with the highest or lowest value. We can use explode to emphasize any number of wedges of the pie. explode allows us to specify an offset value for each category.

```python
import matplotlib.pyplot as plt
import numpy as np

# creating a list of students and their marks in two different subjects 
student = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
sub1 = [95, 45, 60, 20, 75, 35]

# setting offset value for marks of each student, where 0.5 is for lowest marks
explode = [0,0,0,0.5,0,0]


fig, axe = plt.subplots(dpi=100)
# adding explode for the minimum value students marks
axe.pie(sub1, labels=student, autopct='%.2f%%', explode=explode, colors=['yellow', 'black', 'grey', 'blue', 'orange', 'red'])

plt.show()
```

## 9. Boxplot
A boxplot is a standardized way of displaying a dataset based on a five-number summary: the minimum, the maximum, the sample median, and the first and third quartiles. In statistics, the boxplot is a method for graphically depicting groups of numerical data through their quartiles.


A box plot shows the distribution of data along a single axis, using a "box" and "whiskers". The lower end of the box represents the 1st quartile (i.e. 25% of values are below it), and the upper end of the box represents the 3rd quartile (i.e. 25% of values are above it). The median value is represented via a line inside the box. The "whiskers" represent the minimum & maximum values (sometimes excluding outliers, which are represented as dots).



<img align="right" width="380" height="100"  src="images/boxplot.png"  >

In Matplotlib, `boxplot()` is the core function to create box plot. Below are some of the important parameters of the function:

- `x`: Array or a sequence of vectors. The input data.
- `vert`: Set as True or False. The default value is True, which displays the boxes vertically.
- `labels`: Sets the labels for each dataset.
- `notch`: Set as True or False. the default value is False. If True, the parameter will produce a notched box plot.
- `widths`: Sets the width of the box.
- `patch_artist`: Set as True or False. the default value is False. If False, the parameter will produce boxes with the Line2D artist. Otherwise, the boxes will be drawn with Patch artists.

## 10. Stem Plot
stem plots are similar to scatter plots because there are no links between individual data points. However, stem plots include vertical lines that connect each data point to the baseline, while placing a marker at the tip. The stem plot has the advantage of being able to clearly show the shape of a data set’s distribution.

Below are some important parameters required by stem().

- `x`: The position of the stems on the x-axis.
- `y`: The position of the stem heads on the y-axis.
- `linefmt`: A string defining the properties of the vertical lines.
- `markerfmt`: A string defining the properties of the markers at the stem heads.
- `bottom`: The position of the baseline on the y-axis

```python
import matplotlib.pyplot as plt
import numpy as np

# creating dataset
x = np.linspace(0.1, 2 * np.pi, 31)
y = np.exp(np.sin(x))

# Plotting Stem plot
plt.stem(x, y)
plt.show()
```

## 11. Violin Plot
Violin plots are similar to box plots, except that they also show the probability density of the data at different values. These plots include a marker for the median of the data and a box indicating the interquartile range, as in the standard box plots.

A violin plot is more informative than a plain box plot. In fact while a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data.

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1,
               palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
```

## 13. Radar Plot
The radar chart, is also known as web chart, spider chart, or spider web chart, is a chart that consists of a sequence of equi-angular spokes, called radii, where each spoke represents one of the variables that is impacting our data. The data length of a spoke is proportional to the magnitude of the variable for that data point in relation to the maximum magnitude of the variable across all data points. The radar chart is useful in showing relative values for a single data point, or for comparing two or more items as they relate to various categories.

- `Step 1`: Calculate the angle for the value of each category. We begin by assigning each category value an angle. The angles are accumulated according to the order in which they occur.
- `Step 2`: Append the first item to the last, making the circular graph close.
- `Step 3`: Set the coordinate system as polar for your figure.
- `Step 4`: Use the plot() function to plot. We use plot() here in the same way that we used it in the lesson How to Draw a Line Plot.
- `Step 5`: Fill the area. This step is optional.

```python
import matplotlib.pyplot as plt
import math

# creating Dataset
labels = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
values = [95, 45, 60, 20, 75, 35]

# Append the first item to the last, making the circular graph close
values += values[:1]

# Calculate the angle for the value of each category
angles = [n / float(len(labels)) * 2 * math.pi for n in range(len(labels))]
angles += angles[:1]

#  Set the coordinate system as polar
fig, axe = plt.subplots(subplot_kw=dict(polar=True), dpi=100)

# setting ticks
axe.set_xticks(angles[:-1])

# setting tick labesl
axe.set_xticklabels(labels, color='r')

# Plotting Radar between angles and values
axe.plot(angles, values)
axe.fill(angles, values, 'skyblue', alpha=0.4)

plt.show()
```
